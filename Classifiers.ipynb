```python
#DECISION TREE
#import pandas library
import pandas as pd
```

```python
#loading dataset
df=pd.read_csv("diabetes.csv")
df.head()
```

```python
#feature variables
x=df.drop(['Outcome'], axis=1)
x
```

```python
#target variable
y=df.Outcome
y
```

```python
from sklearn.tree import DecisionTreeClassifier # Import Decision Tree Classifier
from sklearn.model_selection import train_test_split # Import train_test_split function
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=1)
```

```python
# Create Decision Tree classifer object
model = DecisionTreeClassifier()

# Train Decision Tree Classifer
model = model.fit(x_train,y_train)

#Predict the response for test dataset
y_pred = model.predict(x_test)
```

```python
#Evaluation using Accuracy score
from sklearn import metrics #Import scikit-learn metrics module for accuracy calculation
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
print("Precision:",metrics.precision_score(y_test,y_pred))
print("Recall:",metrics.recall_score(y_test,y_pred))
print("f1 score:",metrics.f1_score(y_test, y_pred))
```

```python
#Evaluation using Confusion matrix
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test,y_pred)
```

```python
#Importing testing data
testing = pd.read_csv('diabetes.csv')
X_test = testing.iloc[:, :1].values
y_test = testing.iloc[:, 4:5].values
```

```python
pima.head()
```

```python
from sklearn.ensemble import RandomForestRegressor
regressor = RandomForestRegressor(n_estimators = 100, random_state = 0,min_samples_leaf=1)
regressor.fit(X_train, y_train)
y_pred = regressor.predict(y_test)
```

```python
plt.rcParams["figure.figsize"] = (20,8)



plt.plot(X_train, y_train, color = 'red', label="Actual Train")
plt.plot(X_test,y_test, color='green', label="Actual Test")
plt.plot(X_test,y_pred, color='blue', label="Prediction")

plt.title('Diabetes')
plt.xlabel('Age')
plt.ylabel('Outcome')
plt.legend()
plt.show()


```
